```python
df=df.dropna(subset="City")
df
```

```python
#Checking for missing Value
df.isnull().sum()
#For city because it is categorical data(not numerical) we cant implemet mean so thats why we removed it ,for items_count we may can.
```

```python
#Filling missing value in Items_Count
x=df["Items_Count"].mode()
x[0]
df["Items_Count"].fillna(x[0],inplace=True)
```

```python
df.isnull().sum()
```

Filling NA values using group by mean

```python
df["Customer_Rating"]=df.groupby("Company")["Customer_Rating"].transform(lambda x:x.fillna(x.mean()))
df.isnull().sum()
```

```python
df["Delivery_Partner_Rating"]=df.groupby("Delivery_Time_Min")["Delivery_Partner_Rating"].transform(lambda y:y.fillna(y.mean()))
```

```python
df.isnull().sum()
```

```python
df["Delivery_Partner_Rating"].fillna(df["Delivery_Partner_Rating"].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

8) Which Company has the Best operation Efiicieny (Deliver time vs Order Volume)?

```python
company_efficiency=df.groupby("Company").agg(Total_orders=("Order_ID","count"),Avg_Delivery_time=("Delivery_Time_Min","mean")).reset_index()
company_efficiency
```

```python
#Normalizing the company Efiieicing df ,coz We use normalization to put all numeric values on the same scale so no variable dominates just because it’s bigger.
! pip install scikit-learn
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Select coloumns for scaling 
scaler=MinMaxScaler() #Create a scaler
company_efficiency[["Total_orders_Scaled","Avg_Delivery_time_Scaled"]]=scaler.fit_transform(company_efficiency[["Total_orders","Avg_Delivery_time"]])
```

```python
company_efficiency["Efiiciency"]=company_efficiency['Total_orders_Scaled']-company_efficiency["Avg_Delivery_time_Scaled"]
company_efficiency=company_efficiency.sort_values("Efiiciency",ascending=False)
company_efficiency
```

```python
plt.bar(company_efficiency['Company'],company_efficiency['Efiiciency'])
plt.xticks(rotation=60)
plt.xlabel("Company")
plt.ylabel("Efiiciency")
plt.title("Operation Efficieny of Copmany")
plt.grid(True)
plt.show()
```

```python
fig=px.scatter(company_efficiency,x="Total_orders",y="Avg_Delivery_time",color="Company",text="Company",size="bubble_size",size_max=40,title="Operation Efficiency By Companies")
#Creating Buuble Size
minn_eff=company_efficiency['Efiiciency'].min()
company_efficiency["bubble_size"]=company_efficiency['Efiiciency']-minn_eff + 0.1
fig
```
